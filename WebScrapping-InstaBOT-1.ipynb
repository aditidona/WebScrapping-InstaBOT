##### SEARCH_HANLDES: Function to search text on the search bar and open the drop down. (VERY USEFUL IN PARTS OF THE PROJECT)

```python
def search_handles(text):
    search=driver.find_element_by_class_name('wUAXj')
    search.click()
    t=driver.find_element_by_class_name('x3qfX')
    t.send_keys(text)
    t.send_keys(Keys.ENTER)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'fuqBx')))
```

## 4.Follow/Unfollow given handle - 
##### 4.1) Open the Instagram Handle of “So Delhi”
##### 4.2) Start following it. Print a message if you are already following
##### 4.3) After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def follow_handles(handle_name):
    search_handles(handle_name) #function call
    profile=driver.find_element_by_partial_link_text(handle_name) ## locating the profile in the dropdown
    profile.click() 
    wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers'))) #WAIT UNTIL PAGE LOADS
    
    #FOLLOWING THE PROFILE
    follow=driver.find_element_by_xpath('//button[text()="Follow"]') # locating the follow button
    follow.click() 
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'glyphsSpriteFriend_Follow'))) # wait until page is followed
    print('Followed',handle_name)
    driver.back()

def unfollow_handles(handle_name):
    search_handles(handle_name) #function call
    profile=driver.find_element_by_partial_link_text(handle_name) ## locating the profile in the dropdown
    profile.click() 
    wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers'))) #WAIT UNTIL PAGE LOADS
    
    #UNFOLLOWING PROFILE
    temp=driver.find_element_by_class_name("glyphsSpriteFriend_Follow") #locate button next to 'message' to unfollow account
    temp.click()
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'-Cab_'))) #wait for the options to load
    unfollow=driver.find_element_by_class_name('-Cab_') #locate the unfollow option
    unfollow.click()
    print('Unfollowed',handle_name)
    driver.back()
    
```

```python
follow_handles('sodelhi')
unfollow_handles('sodelhi')
```

## 5. Like/Unlike posts
##### 5.1) Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
def like_first_30_posts(handle_name):
    open_profile(handle_name)
    #LIKING
    posts=driver.find_elements_by_class_name('_9AhH0') # locating all the posts
    for i in range(30):
        posts[i].click()    # click on the post
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_97aPb'))) #WAIT FOR POST DIALOG TO OPEN
        like=driver.find_element_by_xpath('//section/span[@class="fr66n"]/button') #locate the like button
        like.click()  
        print('POST',i+1,'LIKED')
        cross=driver.find_element_by_css_selector("[aria-label=Close]") #locate the cross button and close the post
        cross.click()
        wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers'))) #WAIT UNTIL WEBPAGE LOADS
    print('FIRST 30 POSTS OF',handle_name,'LIKED')    
```

```python
like_first_30_posts('dilsefoodie')
```

##### 5.2) Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
def unlike_first_30_posts(handle_name):
    open_profile(handle_name)
    #LIKING AND UNLIKING
    posts=driver.find_elements_by_class_name('_9AhH0') # locating all the posts
    for i in range(30):
        posts[i].click()    # click on the post
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_97aPb'))) #WAIT FOR POST DIALOG TO OPEN
        like=driver.find_element_by_xpath('//section/span[@class="fr66n"]/button') #locate the like button
        like.click()  
        print('POST',i+1,'UNLIKED')
        cross=driver.find_element_by_css_selector("[aria-label=Close]") #locate the cross button and close the post
        cross.click()
        wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers'))) #WAIT UNTIL WEBPAGE LOADS
    print('FIRST 30 POSTS OF',handle_name,'UNLIKED')
```

## 6. Extract list of followers
##### 6.1) Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

```python
def first_500_followers(handle_name):
    open_profile(handle_name)
    follow=driver.find_element_by_partial_link_text('followers') #locating followers hyperlink
    follow.click() # clicking to open list of followers
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,"PZuss"))) #wait until list opens
    #SCROLL TO 500th FOLLOWER
    scroll=0
    while scroll<45:
        cur=driver.find_element_by_xpath('//div[@class="PZuss"]/li[last()]')
        driver.execute_script("arguments[0].scrollIntoView(true);",cur)
        time.sleep(2)
        scroll=scroll+1
    
    #FETCH ALL THE USERNAMES 
    usernames=driver.find_elements_by_class_name('notranslate') 
    followers=[]
    for i in range(500):
        followers.append(usernames[i].text)
    cross=driver.find_element_by_css_selector("[aria-label=Close]") #locate the cross button to close the post dialog
    cross.click()
    wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers'))) #WAIT UNTIL WEBPAGE LOADS
    home=driver.find_element_by_css_selector("[aria-label=Home]")  #back to timeline
    home.click()
    print('------FOLLOWERS EXTRACTED------')
    
    return followers
```

## 7.Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
##### 7.1) If You have already seen the story.
##### 7.2) Or The user has no story.
##### 7.3) Or View the story if not yet seen.

```python
def story_status(handle_name):
    open_profile(handle_name) #function call
    data=bs(driver.page_source,'html.parser')
    story=data.find('canvas',class_='CfWVH') #locate 'canvas' tag
    
    #IF THERE IS NO CANVAS TAG MEANS THE PROFILE DOESNT HAVE A STORY
    if(story==None):
        print('USER HAS NO STORY')
        
    #OTHERWISE IT DOES HAVE A STORY
    else:
        width=int(story['width']) 
        if width==109: #IF THE STORY IS SEEN THE WIDTH WILL BE '109'
            print('I HAVE ALREADY SEEN THE STORY')
        elif width==210:  #IF THE STORY IS NOT SEEN YET THE WIDTH WILL BE '210'
            print('STORY NOT VIEWED')
            open_story=driver.find_element_by_class_name('h5uC0') #SO CLICK AND VIEW THE STORY
            open_story.click()
            print('-----CLICKED AND VIEWED THE STORY-----')
```

```python
story_status('coding.ninjas')
```
